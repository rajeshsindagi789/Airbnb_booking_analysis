##**UNDERSTAND THE GIVEN VARIABLES**

**Listing_id :-** This is a unique identifier for each listing in the dataset.

**Listing_name :-** This is the name or title of the listing, as it appears on the Airbnb website.

**Host_id :-** This is a unique identifier for each host in the dataset.

**Host_name :-** This is the name of the host as it appears on the Airbnb website.

**Neighbourhood_group :-** This is a grouping of neighborhoods in New York City, such as Manhattan or Brooklyn.

**Neighbourhood :-** This is the specific neighborhood in which the listing is located.

**Latitude :-** This is the geographic latitude of the listing.

**Longitude :-** This is the geographic longitude of the listing.

**Room_type :-** This is the type of room or property being offered, such as an entire home, private room, shared room.

**Price :-** This is the nightly price for the listing, in US dollars.

**Minimum_nights :-** This is the minimum number of nights that a guest must stay at the listing.

**Total_reviews :-** This is the total number of reviews that the listing has received.

**Reviews_per_month :-** This is the average number of reviews that the listing receives per month.

**Host_listings_count :-** This is the total number of listings that the host has on Airbnb.

**Availability_365 :-** This is the number of days in the next 365 days that the listing is available for booking.

### **10. What factors are correlated with the prices of Airbnb rentals in New York City? Are variables such as the number of reviews, neighbourhood_group and neighbourhood.**

```python
# Select relevant columns for analysis
columns = ['price', 'number_of_reviews', 'neighbourhood_group', 'neighbourhood']

# Remove rows with missing values in selected columns
data_subset = Data_frame[columns].dropna()

# Convert the categorical variables to numerical using label encoding
data_subset['neighbourhood_group'] = pd.factorize(data_subset['neighbourhood_group'])[0]
data_subset['neighbourhood'] = pd.factorize(data_subset['neighbourhood'])[0]

# Calculate the correlation matrix
correlation_matrix = data_subset.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

### **Observation :**
The correlation heatmap reveals the following insights regarding the factors correlated with Airbnb rental prices in New York City:

Number of Reviews: There is a weak negative correlation between the number of reviews and prices. This suggests that properties with higher prices tend to have fewer reviews, indicating that they may be less frequently booked or targeted towards a specific clientele.

Neighbourhood Group: The neighbourhood_group variable shows a modest correlation with prices. Certain neighbourhood groups may have higher average prices compared to others, indicating variations in demand and desirability across different areas of New York City.

Neighbourhood: The neighbourhood variable also exhibits a modest correlation with prices. Specific neighbourhoods within each neighbourhood group may command higher or lower prices based on their location, amenities, and popularity among visitors

### **Conclusion :**
The correlation analysis reveals that the number of reviews, neighbourhood group, and neighbourhood have a weak association with Airbnb rental prices in New York City. While these factors may have some influence on pricing, they are not the sole determinants. Additional variables and factors should be considered when assessing and predicting rental prices in the city.
